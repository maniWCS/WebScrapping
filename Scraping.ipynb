### Bypassing the protection by specifying an existing browser

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

headers = {
    'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10.12; rv:55.0) Gecko/20100101 Firefox/55.0 Chrome/91.0.4472.114',
}

page = requests.get('https://chucknorrisfacts.net/facts.php?page=1', headers=headers)
soup = BeautifulSoup(page.content, 'html.parser')

divs = soup.find_all('div', attrs={'style': 'border-top: 1px solid #251836; padding: 0 0 5px 7px;'})
```

### How many jokes are there?

```python
print('Total Number = ' , len(divs))
```

### Displays the 8th joke using the .text attribute. Use the .find method with the appropriate tag to isolate only the 8th joke, without the other elements

```python
print('8th joke:', divs[7].find('p').text)
print('8th note:', divs[7].find('span').text)
```

### Create an empty dictionary, then create a loop that will record in this dictionary each joke as a key and each corresponding note as a value

```python
chuck_dictionary ={}

for div in divs:
    chuck_dictionary [div.find('p').text] =div.find('span').text

print(chuck_dictionary)
```

### Transform this dictionary into a DataFrame with 2 columns: joke and note. It must have 20 rows: one per joke.

```python
dfObj = pd.DataFrame(list(chuck_dictionary.items()))
```
